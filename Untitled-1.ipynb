```python
 #plcing dummy variables so as to make qualitative explanatory variables workable in our model
tips['time'] = tips['time'].map({'Dinner': 0, 'Lunch': 1})
tips['sex'] = tips['sex'].map({'Male':0, 'Female':1})
tips['smoker'] = tips['smoker'].map({'Yes': 0, 'No':1})
tips['day'] = tips['day'].map({'Sat':0, 'Sun':1, 'Fri': 2, 'Thur': 3})
tips.head()
```

```python
# proofing if our dummy variables are representive
tips['day'].value_counts()
#it worked
```

```python
# modelling
x = tips.drop('day', axis=1)
y = tips['day']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=40)
from sklearn.preprocessing import StandardScaler
mx = StandardScaler()
x_train = mx.fit_transform(x_train)
x_test = mx.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
pred = lr.predict(x_test)
score = lr.score(x_test, y_test)
print(score)
```

```python
df = pd.DataFrame({pred: 'predicted', y_test:'actual'})
print(df.head())
```
